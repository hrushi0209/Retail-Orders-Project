```python
# read data from files
import pandas as pd
df = pd.read_csv('orders.csv')
df.head(10)
```

```python
#distinct values available in ship mode column
df['Ship Mode'].unique()
```

```python
# Treating "Not Available" and "Unknown" values as NAN
df = pd.read_csv('orders.csv',na_values=["Not Available","unknown"])
df['Ship Mode'].unique()
```

```python
# All columns from dataset
df.columns
```

```python
# Renaming column names, replacing them with lower case 
df.columns = df.columns.str.lower()
df.columns
```

```python
# Replacing spaces with underscore
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df.head(5)
```

```python
# Derive new columns discount 
df['discount'] = df['list_price']*df['discount_percent']*.01
df
```

```python
# Derive new Sales Price
df['sale_price'] = df['list_price']-df['discount']
df
```

```python
# derive the new column total profit
df['total_profit'] = df['sale_price'] - df['cost_price']
df.head(10)
```

```python
#datatypes of all columns
df.dtypes
```

```python
# convert order_date from object datatype to datetime
df['order_date']= pd.to_datetime(df['order_date'],format = "%Y-%m-%d")
```

```python
df.dtypes
```

```python
# drop cost_price,list_price and discount percent columns
df.drop(columns=['list_price','cost_price','discount_percent'],inplace = True)
```
